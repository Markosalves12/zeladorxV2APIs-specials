```python
# exemplo de envio de boby para criar uma nova unidades
new_body = {
     'nome': 'unidade jupyter',
     'linkmapajardinagem': '', # esse campo so sera salvo cao o usuario tenha o macro servico liberado em api/zeladorx/MyMacroServices/
     'linkmapalimnpezapredial': '', # esse campo so sera salvo cao o usuario tenha o macro servico liberado em api/zeladorx/MyMacroServices/
     'empresasecundaria': 6, # api/empresas_secundarias/ListEmpresaSecundariaJardinagemFromForms/
}
# codigo de permissão necessário --> 340: Pode criar novas unidades
response = requests.post("http://127.0.0.1:9000/api/unidades/CreateUnidade/", new_body, headers=headers)
response.json()
```

# Atualizar unidade

```python
# as chaves body da atualização é exatamente igual o body da criacao
# atualiando a nome para unidade jupyter
new_body = {
     'nome': 'unidade jupyter',
     'linkmapajardinagem': '', # esse campo so sera salvo cao o usuario tenha o macro servico liberado em api/zeladorx/MyMacroServices/
     'linkmapalimnpezapredial': '', # esse campo so sera salvo cao o usuario tenha o macro servico liberado em api/zeladorx/MyMacroServices/
     'empresasecundaria': 6, # api/empresas_secundarias/ListEmpresaSecundariaJardinagemFromForms/
}
# codigo de permissão necessário --> 341: Pode editar unidades
response = requests.put("http://127.0.0.1:9000/api/unidades/UnidadeUpdate/hDVjIq1JYBOZ/", new_body, headers=headers)
response.json()
```

# Detalhes unidades

```python
# essa rota recebe o id random da unidade como parametro de url, consultar em api/unidades/ListUnidade/
# detalhes da unidade riada anteriormente, obeserve o id_random
# http://127.0.0.1:9000/api/unidades/UnidadeDetail/id_random/
# codigo de permissão necessário -->  342: Pode visualizar unidades
response = requests.get("http://127.0.0.1:9000/api/unidades/UnidadeDetail/hDVjIq1JYBOZ/", new_body, headers=headers)
response.json()
```

# Alterar status

```python
# essa rota recebe o id random da unidade como parametro de url, consultar em api/unidades/ListUnidade/
# atualizando a unidade criada anteriormente, obeserve o id_random
# codigo de permissão necessário --> 
# 344: Pode desmobilizar unidades ou demosbilizar (Desmobilizado) ou
# 345: Pode reabilitar unidades para reabilitar (Mobilizado)

new_body = {
    'status': 'Desmobilizado'
}

response = requests.put("http://127.0.0.1:9000/api/unidades/UnidadeAlterStatus/hDVjIq1JYBOZ/", new_body, headers=headers)
response.json()
```

# If delete unidade

```python
# essa rota recebe o id random da unidade como parametro de url, consultar em api/unidades/ListUnidade/
# codigo de permissão necessário --> 343: Pode excluir unidades

response = requests.get("http://127.0.0.1:9000/api/unidades/IfDeleteUnidade/hDVjIq1JYBOZ/", headers=headers)
response.json()
```

```python
# essa rota recebe o id random da unidade como parametro de url, consultar em api/unidades/ListUnidade/
# codigo de permissão necessário --> 343: Pode excluir unidades

response = requests.delete("http://127.0.0.1:9000/api/unidades/DeleteUnidade/75X3nHLHlXKc/", headers=headers)
response.json()
```
