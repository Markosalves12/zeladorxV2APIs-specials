# Troca de senha

```python
body = {
    "email": username
}

response = requests.post(
    "http://127.0.0.1:9000/api/reset-password/",
    body   # <<< aqui o correto é json, não body direto
)
# o token de troca e enviado por emial, porem o token de url fica visivel
```
