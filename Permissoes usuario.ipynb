# Permissoes dos usuário

```python
# retorna o estsado atual das permissoes de todos os usuarios que é permitido visualizar
# vem de api/gerentes/ListGerenteJardinagem/
# codigo de permissao --> 301: Pode visualizar permissões de limpeza predial
response = requests.get("http://127.0.0.1:9000/api/permissions/ListPermissionsAccessEspecials/", headers=headers)
response.json()['results']
df = pd.DataFrame(response.json()['results'])
display(df)
```

# Permissoes por usuario

```python
# recebe o id_random da permissao como parametro de url
# retorna as permissoes de um usuario especifico, sendo identificado pelo id_random
# codigo de permissão necessário --> 301: Pode visualizar permissões de jardinagem
response = requests.get("http://127.0.0.1:9000/api/permissions/PermissionsAccessEspecialsDetail/nbkpBEoDuW4Z/", headers=headers)
df = pd.DataFrame(response.json())
display(df)
```

# atualizar permissoes

```python
body = {
    'Gerente': 2,  # vem de api/gerentes/ListGerenteLimpezaPredial/
    'Permissions': [1, 2, 3]  # api/permissions/ListPermissionsLimpezaPredial/
}
```

```python
# atualiza o cinjunto de permissoe s de um usuario ListPermissionsAccessJardinagem
# codigo de permissao --> 300: Pode editar permissões de jardinagem
response = requests.put("http://127.0.0.1:9000/api/permissions/PermissionsAccessEspecialsUpdate/nbkpBEoDuW4Z/", body, headers=headers)
response.json()
```
